## Formatting the API
The mess of outputs needs to be converted before it can be seen properly. This is done here with this code, which lists each search result by their author, thumbnail, title, views, and length

```python
# Procedural abstraction of Print code for coins
def print_result(v):
    print("Result Type: ", v["type"])
    print("Author: " + v["video"]["author"]["title"])
    print("Thumbnail: " + v["video"]["thumbnails"][0]["url"])
    print("Title: " + v["video"]["title"])
    views=v["video"]["stats"]["views"]
    if views>1000000000:
        print("Views: "+str(views//1000000000)+'B')
    elif views>1000000:
        print("Views: "+str(views//1000000)+'M')
    elif views>1000:
        print("Views: "+str(views//100)+'K')
    else:
        print("Views: "+str(views))
    length=v["video"]["lengthSeconds"]
    if length%60<10:
        print("Length: "+str(length//60)+':0'+str(length%60))
    else:
        print("Length: "+str(length//60)+':'+str(length%60))
    
try:
    print("JSON data is Python type: " + str(type(json))) # Test if json is processed successfully
    try:
        print('')
        # For each video in the json, view the results using print_results
        for video in json['contents']:
            print_result(video)
            print()
            
    except:
        print("Did you insert a valid key in X-RapidAPI-Key of API cell above?")
        print(json)
except:
    print("This cell is dependent on running API call in cell above!")
```
